# Relation between variables 

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 


sns.scatterplot(x='5.1', y='3.5', 
                hue='Iris-setosa', data=df, ) 

# Placing Legend outside the Figure 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 

plt.show()
```

# Comparing two variables

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 


sns.scatterplot(x='1.4', y='0.2', 
                hue='Iris-setosa', data=df, ) 

# Placing Legend outside the Figure 
plt.legend(bbox_to_anchor=(1, 1), loc=2) 

plt.show()
```

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# Assuming 'Iris-setosa' is the column name
sns.pairplot(df.drop(['5.1'], axis=1),  
             hue='Iris-setosa', height=2)
plt.show()  # Don't forget to show the plot
```

# Histogram

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 


fig, axes = plt.subplots(2, 2, figsize=(10,10)) 

axes[0,0].set_title("5.1") 
axes[0,0].hist(df['5.1'], bins=7) 

axes[0,1].set_title("3.5") 
axes[0,1].hist(df['3.5'], bins=5); 

axes[1,0].set_title("1.4") 
axes[1,0].hist(df['1.4'], bins=6); 

axes[1,1].set_title("0.2") 
axes[1,1].hist(df['0.2'], bins=6);
```

# Histograms with Distplot Plot

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

plot = sns.FacetGrid(df, hue="Iris-setosa") 
plot.map(sns.distplot, "5.1").add_legend() 

plot = sns.FacetGrid(df, hue="Iris-setosa") 
plot.map(sns.distplot, "3.5").add_legend() 

plot = sns.FacetGrid(df, hue="Iris-setosa") 
plot.map(sns.distplot, "1.4").add_legend() 

plot = sns.FacetGrid(df, hue="Iris-setosa") 
plot.map(sns.distplot, "0.2").add_legend() 

plt.show()
```

# Heatmap

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# Assuming 'Iris-setosa' is a non-numeric column that needs to be excluded
numeric_df = df.drop(columns=['Iris-setosa'], errors='ignore')

# Generate heatmap of the correlation matrix
sns.heatmap(numeric_df.corr(method='pearson'), annot=True)
plt.show()

```

# Box Plot

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

def graph(y): 
	sns.boxplot(x="Iris-setosa", y=y, data=df) 

plt.figure(figsize=(10,10)) 
	
# Adding the subplot at the specified 
# grid position 
plt.subplot(221) 
graph('5.1') 

plt.subplot(222) 
graph('3.5') 

plt.subplot(223) 
graph('1.4') 

plt.subplot(224) 
graph('0.2') 

plt.show()
```
